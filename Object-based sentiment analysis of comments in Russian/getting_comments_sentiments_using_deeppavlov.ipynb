```python
import json
import requests
from requests.structures import CaseInsensitiveDict
import csv
import ssl
import json

try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context

comts = 'comments.csv'
```

```python
def read_and_prepare(file):
    comments_file = open(file)
    comments = csv.reader(comments_file, delimiter=';')
    tpl = []

    for i in comments:
        tpl.append(i)

    flat_list = []
    for sublist in tpl:
        for item in sublist:
            if item != ' ' and item != '':
                flat_list.append(item)

    comments_file.close()

    return flat_list
```

```python
user_comments = read_and_prepare(comts)
```

```python
## I used a local Docker version of DeepPavlov for this analysis
url = "http://127.0.0.1:5555/model"
headers = CaseInsensitiveDict()
headers["Content-Type"] = "application/json"
headers["Accept"] = "application/json"
sentiments = []
```

```python
for comment in user_comments:
    text = comment.replace('"', '')
    text = text.replace('\\', '')
    data = '{"x": ["' + text + '"]}'
    resp = requests.post(url, headers=headers, data=data.encode('utf-8'))
    sentiment = json.loads(resp.text)
    print(sentiment)
    sentiments.append(sentiment[0][0])
```

```python
print(len(user_comments))
print(len(sentiments))
```

```python
amount_of_positive = 0
amount_of_negative = 0
amount_of_neutral = 0

for result in sentiments:
    if result == 'positive':
        amount_of_positive += 1
    if result == 'negative':
        amount_of_negative += 1
    if result == 'neutral' or result == 'speech' or result == 'skip':
        amount_of_neutral += 1

print('positive  ', amount_of_positive)
print('negative  ', amount_of_negative)
print('neutral  ', amount_of_neutral)
```

```python
sentiments_file = open('deeppavlov_sent.txt', 'w')

for info in sentiments:
    sentiments_file.write(info)
    sentiments_file.write('\n')

sentiments_file.close()
```
